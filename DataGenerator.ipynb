# Bezier Distortion

```python
import numpy as np
from PIL import Image
from scipy.interpolate import interp2d
import matplotlib.pyplot as plt

def apply_bezier_distortion(image, intensity=10, grid_size=4):
    """
    Apply a controlled elastic distortion using Bézier curves on a grid.
    
    :param image: Input image (PIL Image object).
    :param intensity: Intensity of the distortion.
    :param grid_size: Size of the grid (number of grid points).
    :return: Distorted image (PIL Image object).
    """
    # Convert PIL Image to NumPy array
    img_array = np.array(image)
    rows, cols, ch = img_array.shape

    # Create a meshgrid for pixel coordinates
    src_x, src_y = np.meshgrid(np.linspace(0, cols-1, num=cols), np.linspace(0, rows-1, num=rows))
    dst_x, dst_y = src_x.copy(), src_y.copy()

    # Generate control points for the grid
    control_points = np.linspace(0, cols, num=grid_size)
    displacement = np.random.normal(size=(grid_size, grid_size, 2), scale=intensity)

    # Interpolate displacements using Bézier curves (spline interpolation)
    for i in range(ch):
        fx = interp2d(control_points, control_points, displacement[:, :, 0], kind='cubic')
        fy = interp2d(control_points, control_points, displacement[:, :, 1], kind='cubic')
        dst_x += fx(np.arange(cols), np.arange(rows))
        dst_y += fy(np.arange(cols), np.arange(rows))

    # Map coordinates using interpolation
    map_x = np.clip(dst_x, 0, cols - 1).astype(np.float32)
    map_y = np.clip(dst_y, 0, rows - 1).astype(np.float32)
    distorted_array = np.zeros_like(img_array)

    for i in range(ch):
        interp_func = interp2d(src_x[0, :], src_y[:, 0], img_array[:, :, i], kind='cubic')
        distorted_array[:, :, i] = interp_func(map_x[0, :], map_y[:, 0])

    # Convert back to PIL Image
    distorted_image = Image.fromarray(np.uint8(distorted_array))
    return distorted_image

# Example usage
image = Image.open('Data/Dataset/word_1.png')  # Replace with the path to your image
distorted_image = apply_bezier_distortion(image, intensity=0.005, grid_size=4)
distorted_image.show()
```

# Multi Scale Distortion

```python
import cv2
import numpy as np
from PIL import Image

def apply_multi_scale_distortion(image, scale_factor):
    """
    Apply multi-scale distortion to an image using OpenCV and convert between PIL and OpenCV.
    
    :param image: Input image (PIL Image object).
    :param scale_factor: Scaling factor (less than 1 to reduce, more than 1 to enlarge).
    :return: Scaled image (PIL Image object).
    """
    # Convert PIL Image to NumPy array (OpenCV format)
    open_cv_image = np.array(image)
    # Convert RGB to BGR (OpenCV uses BGR)
    open_cv_image = open_cv_image[:, :, ::-1].copy()

    # Calculate new size based on scale factor
    new_size = (int(open_cv_image.shape[1] * scale_factor), int(open_cv_image.shape[0] * scale_factor))
    # Resize the image using OpenCV
    resized_image = cv2.resize(open_cv_image, new_size, interpolation=cv2.INTER_LINEAR)

    # Convert back to RGB from BGR
    resized_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
    # Convert NumPy array back to PIL Image
    pil_image = Image.fromarray(resized_image)
    
    
    return pil_image
```

# Elastic Distortion

```python
import numpy as np
from PIL import Image
import scipy.ndimage

def elastic_transform(image, alpha, sigma, random_state=None):
    """
    Apply elastic distortion on an image.
    
    :param image: Input image (PIL Image object).
    :param alpha: Scale of transformation. This is the intensity of the deformation.
    :param sigma: Standard deviation of the Gaussian filter. This controls the smoothness of the deformation.
    :param random_state: Random state for reproducibility.
    :return: Distorted image (PIL Image object).
    """
    if random_state is None:
        random_state = np.random.RandomState(None)

    # Convert PIL Image to numpy array
    image_array = np.array(image)

    # Shapes
    shape = image_array.shape
    dx = scipy.ndimage.gaussian_filter((random_state.rand(*shape) - 0.5) * 2, 
                                       sigma, mode="constant", cval=0) * alpha
    dy = scipy.ndimage.gaussian_filter((random_state.rand(*shape) - 0.5) * 2, 
                                       sigma, mode="constant", cval=0) * alpha

    # Create meshgrid
    x, y, z = np.meshgrid(np.arange(shape[1]), np.arange(shape[0]), np.arange(shape[2]))

    # Indices
    indices = np.reshape(y+dy, (-1, 1)), np.reshape(x+dx, (-1, 1)), np.reshape(z, (-1, 1))

    # Map coordinates from the distorted image to the original image
    distorted_image = scipy.ndimage.map_coordinates(image_array, indices, order=1, mode='reflect')

    # Convert the distorted array back to PIL Image
    distorted_image = Image.fromarray(distorted_image.reshape(shape).astype(np.uint8))

    return distorted_image

# Load your image (make sure to provide the correct path)
image_path = 'Data/Dataset/word_9.png'  # Update with the actual path
image = Image.open(image_path)

# Parameters for the transformation
alpha = 1000 # Intensity of the deformation
sigma = 11  # Smoothness of the deformation

# Apply elastic transformation
transformed_image = elastic_transform(image, alpha, sigma)

# Show or save the result
transformed_image.show()
# transformed_image.save('elastic_transformed_image.jpg')
```

# Resizing all the images to fixed size

```python
import os
from PIL import Image

def resize_images_to_height(directory, output_directory, target_height=32):
    """
    Resize all images in the specified directory to a given height, maintaining aspect ratio.
    
    :param directory: Path to the directory containing image files.
    :param output_directory: Path to the directory where resized images will be saved.
    :param target_height: The target height to which images will be resized.
    """
    # Create the output directory if it does not exist
    os.makedirs(output_directory, exist_ok=True)

    # Iterate through all files in the directory
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        
        # Check if the file is an image
        if file_path.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')):
            try:
                with Image.open(file_path) as img:
                    # Calculate the new width to maintain the aspect ratio
                    aspect_ratio = img.width / img.height
                    new_width = int(target_height * aspect_ratio)

                    # Resize the image using the LANCZOS resampling filter
                    resized_img = img.resize((new_width, target_height), Image.Resampling.LANCZOS)

                    # Save the resized image to the output directory
                    resized_img.save(os.path.join(output_directory, filename))

            except IOError:
                print(f"Error processing image {filename}")

# Specify the directory containing your images and the output directory
image_directory = "Dataset"  # Update this to the path of your image folder
output_directory = "final"  # Update this to your desired output folder

# Resize the images
resize_images_to_height(image_directory, output_directory)
print("Images have been resized and saved.")
```
